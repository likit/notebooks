```python
def load_ebseq_results(fname, FDR=0.05):
    de = []
    de_fc = []
    not_de = []
    not_de_fc = []
    for n, line in enumerate(open(fname)):
        if n == 0:
            continue
        gene_name, ppee, ppde, postfc, realfc = line.split()
        ppee = float(ppee) # posterior probability that genes are equally expressed
        ppde = float(ppde) # posterior probability that genes are differentially expressed
        postfc = float(postfc) # the ratio between posterior mean expression estimates of the gene for each condition
        realfc = float(realfc)  # the real fold change (condition 1 over condition 2) for a gene
        log_ppde = (-1) * log10(1.0 - ppde)
        if (1.0 - ppde) <= FDR:
            de.append((log_ppde))
            de_fc.append(log(realfc))
        else:
            not_de.append((log_ppde))
            not_de_fc.append(log(realfc))
    return de, de_fc, not_de, not_de_fc
```
