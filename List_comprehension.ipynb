#List comprehension

##Number example

List comprehension is used to create or manipulate lists. The syntax is simpler and it usually runs faster than for loop. It works with any kind of data structure that for loop can iterate (list, set, file, generator and etc.).

```python
theList = [10,2,31,4,56,6,47,81,90,100]
```

```python
sq_nums = []
for number in theList:
    sq_nums.append(number ** 2)

print sq_nums
```

```python
sq_nums = [x**2 for x in theList]
print sq_nums
```

```python
sqrt_nums = [sqrt(x) for x in theList]
print sqrt_nums
```

```python
import random
random_numbers = [random.random() for i in range(20)]
plot(random_numbers, 'o')
```

```python
numbers = linspace(0, 20, 1000)
plot(numbers)
```

```python
plot([sin(x) for x in numbers])
title('sin(x)')
```

```python
plot([cos(x) for x in numbers], '--', color='red')
title('cos(x)')
```

##String examples

```python
sentence = 'the quick brown fox jumped over the lazy dog'
print sentence.split()
```

```python
# capitalize all words in the sentence
cap_words = []
for word in sentence.split():
    cap_words.append(word.capitalize())
    
print cap_words
```

```python
# capitalize all words in the sentence with comprehension
print [word.capitalize() for word in sentence.split()]
```

```python
# Change case with list comprehension
print [word.upper() for word in sentence.split()]
```

```python
# plot word length distribution using for loop
sentence = 'the quick brown fox jumped over the lazy dog'
word_lens = []
for word in sentence.split():
    word_lens.append(len(word))
    
hist(word_lens)
xticks([3,4,5,6])
yticks([1,2,3,4,5,6])
xlabel('length')
ylabel('number')
title('Word length distribution')
```

```python
# plot word length distribution with comprehension
hist([len(word) for word in sentence.split()])
xlabel('length')
ylabel('number')
xticks([3,4,5,6])
yticks([1,2,3,4,5,6])
title('Word length distribution')
```

##Comprehension with condition

```python
low = []
high = []
for n in theList:
    if n >= 50:
        high.append(n)
    else:
        low.append(n)

print len(low), len(high)
```

```python
print "Total low values =", len([x for x in theList if x < 50]),
print "Total high values =", len([x for x in theList if x >= 50])
```

```python
print [x for x in theList if x % 2 != 0]
```

```python
print [word.capitalize() for word in sentence.split() if len(word) == 4]
```

##Advanced examples

```python
# nested list comprehension
print [x for x in [y for y in theList if y % 2 == 0] if x > 20]
```

```python
# with if else statement
print ['hi' if x >= 50 else 'low' for x in theList]
```

```python
# dictionary comprehension from lists with if else statement
theSamples = ['soil1', 'soil2', 'grass1', 'soil3', 'grass2', 'soil4', 'grass3', 'grass4', 'water1', 'water2']
theDict = {theSamples[theList.index(x)]:'hi' if x >= 50 else 'low' for x in theList}
print theDict
```

#Dictionary comprehension

```python
d = {'a':10, 'b':302, 'c':23, 'd':17, 'e':122, 'f':200}
```

```python
even_d = {}
for key, value in d.iteritems():
    if value % 2 == 0:
        even_d[key] = value

print even_d
```

```python
even_d = {key:value for (key, value) in d.iteritems() if value % 2 == 0}
print even_d
```

#Set comprehension

```python
codons = ('AAA', 'ATG', 'ATG', 'AAA', 'CGT', 'GGT', 'CCC', 'CCC', 'CGT')
print set(codon for codon in codons if 'A' in codon)
print set(codon for codon in codons if 'C' in codon)
```
