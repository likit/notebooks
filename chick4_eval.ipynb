Reverse relationship
--------------------

```python
unmatched_chrs_rev = {}
for gene, chr in unmatched_chrs.iteritems():
    name = gene.split(':')[0]
    if name != chr:
        try:
            unmatched_chrs_rev[name][chr] += 1
        except KeyError:
            if name not in unmatched_chrs_rev:
                unmatched_chrs_rev[name] = {}
                unmatched_chrs_rev[name][chr] = 1
            else:
                if chr not in unmatched_chrs_rev[name]:
                    unmatched_chrs_rev[name][chr] = 1
```

```python
unmatched_chrs_rev_graphs = {}
for chr in unmatched_chrs_rev:
    G = nx.DiGraph()
    G.add_node(chr, {'size':300, 'transcript':0})
    for c, n in unmatched_chrs_rev[chr].iteritems():
        G.add_node(c)
        G.node[c]['size'] = n*10
        G.node[c]['transcript'] = n
        G.node[chr]['transcript'] += n
        G.node[chr]['size'] += n*10
        G.add_edge(chr, c)
    unmatched_chrs_rev_graphs[chr] = G
```

```python
chromosome = 'chrUn_random'
H = unmatched_chrs_rev_graphs[chromosome]
labels = {}
for nd in H.node:
    if H.node[nd]['transcript'] > 0:
        label = '%s (%d)' % (nd, H.node[nd]['transcript'])
    else:
        label = ''
    labels[nd] = label

hub_ego = nx.ego_graph(H, chromosome)
pos = nx.spring_layout(hub_ego)
color = []
for v in H.node:
    transcript = H.node[v]['transcript']
    if transcript < 50:
        c = 2.0
    elif transcript < 100:
        c = 10.0
    elif transcript < 200:
        c = 14.0
    else:
        c = 20.0
    color.append(c)

size = [H.node[n]['size'] for n in H.node]
nx.draw(hub_ego, pos, node_color=color, node_size=size, labels=labels, alpha=0.9, font_size=10, arrows=False, cmap=cm.Blues)
nx.draw_networkx_nodes(hub_ego,pos,nodelist=[chromosome],node_size=13550,node_color='g', alpha=0.9)
axis('off')
savefig('chrUn_random_dist.pdf')
```

Comparison of number of exons and sizes of mapped transcripts
-------------------------------------------------------------
Evaluating deviation of a number of exons and length of each gene.
**Note that some transcripts are mapped to multiple locations on chick4 genome. Only the longest alignment of each transcript is used in this analysis.

```python
from collections import namedtuple
Trns = namedtuple('Trns', ['name', 'exons', 'length']) # a transcript object
chick3_length_exons = {}
chick4_length_exons = {}
length_exons_diff = {}

for line in open('line67.all.global.local.filtered.100.bed'):
    cols = line.strip().split()
    name = cols[3]
    length = sum([int(l) for l in cols[-2].split(',')])
    exons = int(cols[9])
    trns = Trns(name, exons, length)
    if name not in unmatched_chrs and exons > 2:
        chick3_length_exons[name] = trns

for line in open('line67.all.global.local.chick3.filtered.100.chick4.psl.sorted.best.psl'):
    cols = line.strip().split()
    name = cols[9]
    length = sum([int(l) for l in cols[-3].split(',')[:-1]])
    exons = int(cols[-4])
    trns = Trns(name, exons, length)
    if name in unmatched_chrs or exons < 3:
        continue
    else:
        chick4_length_exons[name] = trns
    '''
    if name not in chick4_length_exons:
        chick4_length_exons[name] = trns
    else:
        if length > chick4_length_exons[name].length:
            chick4_length_exons[name] = trns'''

for transcript in chick4_length_exons:
    if transcript in chick3_length_exons:
        trns3 = chick3_length_exons[transcript]
        trns4 = chick4_length_exons[transcript]
        if trns3.length - trns4.length > 100: # ignore subtle difference in size.
            length_exons_diff[transcript] = (trns4.length - trns3.length, trns4.exons - trns3.exons)
```

```python
length_diff_few = [] # fewer exons, shortened length
exons_diff_few = [] # fewer exons
length_diff_more = [] # more exons, shortened length
exons_diff_more = [] # more exons
length_diff_zero = [] # same exons, shortened length
exons_diff_zero = [] # same exons
for length, exons in length_exons_diff.itervalues():
    if exons == 0:
        length_diff_zero.append(length)
        exons_diff_zero.append(exons)
    elif exons < 0:
        length_diff_few.append(length)
        exons_diff_few.append(exons)
    else:
        length_diff_more.append(length)
        exons_diff_more.append(exons)
```

Transcript Span
===============

To investigate a change in intron size, we will compare a span of each transcript on chick3 and chick4 genome.

A basic formula is **span = chromosome start - chromosome stop**.

And again, we ignore all transcripts with fewer than two exons.

```python
chick4_span = {}
for line in open('line67.all.global.local.chick3.filtered.100.chick4.psl.sorted.best.psl'):
    cols = line.strip().split()
    if cols[9] not in length_exons_diff:
        if int(cols[-4]) > 2:
            chick4_span[cols[9]] = int(cols[16]) - int(cols[15])
```

```python
chick3_span = {}
for line in open('line67.all.global.local.filtered.100.bed'):
    cols = line.strip().split()
    span = int(cols[2]) - int(cols[1])
    if int(cols[-3]) > 2:
        chick3_span[cols[3]] = span
```

```python
common_span_chick3 = []
common_span_chick4 = []
common_span_ratio = []
for trns in chick3_span:
    if trns in chick4_span:
        span3 = chick3_span[trns]
        span4 = chick4_span[trns]
        if (span3/span4) > 2.0 or (span4/span3) > 2.0:
            common_span_chick3.append(chick3_span[trns])
            common_span_chick4.append(chick4_span[trns])
        common_span_ratio.append(span3/span4)
```

Gene expression

Objective:: To compare gene expression between two genomes.
To make a fair comparison a set of genes is limited to genes with the same number of exons and length in both chick3 and chick4.

```python
from collections import namedtuple
```

```python
Transcript = namedtuple('Transcript', ['chrom', 'exon_num', 'length']) 
```

```python
chick3 = {}
for line in open('line67.all.global.local.filtered.100.bed'):
    cols = line.strip().split()
    if int(cols[9]) > 2:
        exon_num = int(cols[9])
        tran_size = sum([int(s) for s in cols[-2].split(',')])
        chrom = cols[0]
        name = cols[3]
        tr = Transcript(chrom, exon_num, tran_size)
        chick3[name] = tr
```

Duplicated sequences
--------------------
Looking for genes matched more than one chromosomes.

```python
chick4 = {}
for line in open('line67.all.global.local.chick3.filtered.100.chick4.psl.sorted.best.psl'):
    cols = line.strip().split()
    exon_num = int(cols[-4])
    if exon_num > 2:
        tran_size = sum([int(s) for s in cols[-3].split(',')[:-1]])
        chrom = cols[13]
        name = cols[9]
        chrom_start = int(cols[15])
        chrom_end = int(cols[16])
        try:
            chick4[name].add((chrom, chrom_start, chrom_end))
        except KeyError:
            chick4[name] = set([(chrom, chrom_start, chrom_end)])
```

```python
chimera = [k for k, v in chick4.iteritems() if len(v) > 1] # k = name, v = a list of chromosomes
```

```python
#for name in chimera:
#    print name, chick4[name]
```

```python
chromosomes = []
starts = []
chroms = {'chr1':1}
n = 1
duplicates = set([])
genes = {}
c = 0
colors = []
for name in chimera:
    gene_name = name.split('.')[0]
    if 'chr1' in [k[0] for k in chick4[name]]:
        sorted_list = sorted([k for k in chick4[name]], key=lambda x: x[0], reverse=True)
        for k, start, end in sorted_list:
            id = name+'-'+k
            if id not in duplicates:
                duplicates.add(id)
                if k not in chroms:
                    n += 1
                    chroms[k] = n
                chromosomes.append(chroms[k])
                starts.append(start)
                if gene_name not in genes:
                    c += 1
                    genes[gene_name] = c
                colors.append(genes[gene_name])
```

```python
#starts = [s+10000 for s in starts]
scatter(chromosomes, starts, marker='^', s=100, alpha=1.0, c=colors, zorder=2)
ylabel('chromosome start')
xlim(0, max(chromosomes)+1.0)
yscale('log')
positions = []; labels = []
for k,v in chroms.iteritems():
    positions.append(v)
    labels.append(k)
xticks(positions, labels)
savefig('duplicate.pdf')
```
