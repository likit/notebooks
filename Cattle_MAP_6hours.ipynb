##Combine counts from two lanes

```python
import os
import glob
import csv
import collections
import pandas as pd
from pandas import DataFrame
import itertools
```

```python
def parse_count(filename):
    with open(filename) as infile:
        csvreader = csv.reader(infile, dialect='excel-tab')
        for row in csvreader:
            geneid = row[0].split('.')[-1]
            count = int(row[3])
            yield geneid, count
    
for f in glob.glob('*L001*.sam.counts'):
    total_counts = collections.defaultdict(int)
    for geneid, count in parse_count(f):
        total_counts[geneid] += count
    for geneid, count in parse_count(f.replace('L001', 'L002')):
        total_counts[geneid] += count
    with open(f.replace('_L001', ''), 'w') as outfile:
        for key in sorted(total_counts):
            print >> outfile, '%s\t%d' % (key, total_counts[key])
```

```python
for f in glob.glob('*L00?*.sam.counts'):
    os.remove(f)  # remove files no longer needed
```

###Convert transcript counts to gene counts

```python
trnx2genes = {}
for line in open('Bos_taurus.UMD3.1.73.cdna.all.fa.header'):
    items = line.split()
    trnxid = items[0].lstrip('>')
    geneid = items[3].split(':')[1]
    trnx2genes[trnxid] = geneid
```

```python
for f in glob.glob('*cdna_bowtie2_raw.sam.counts'):
    genes = collections.defaultdict(int)
    for line in open(f):
        trnxid, count = line.split()
        geneid = trnx2genes[trnxid]
        genes[geneid] += int(count)
    outfile = open(f.replace('.counts', '.gene.counts'), 'w')
    for geneid, count in genes.iteritems():
        print >> outfile, '%s\t%d' % (geneid, count)
    outfile.close()
```

```python
count_data = None
for f in glob.glob('*cdna_bowtie2_raw.sam.gene.counts'):
    col_names = ['geneid', 'X' + f.split('_')[0]]  # names in R cannot start with numbers
    if count_data:
        new_count_data = pd.read_table(f, header=None, sep='\t', names=col_names)
        count_data = pd.merge(count_data, new_count_data, on='geneid')
    else:
        count_data = pd.read_table(f, header=None, sep='\t', names=col_names)
```
