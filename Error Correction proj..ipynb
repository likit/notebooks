#SEECER installation and running (HPCC dev-amd09 machine)

##Without quality trimming

Warning: the program will crash if the same temporary directory (-t option) is specified for multiple datasets at the same time.
Unfortunately, the program does not create a specified temporary directory if it does not exist. It has to be created beforehands.

    $ ./bin/run_seecer.sh -t /mnt/scratch/preeyanon/tmp/line7i_notrim/ /mnt/ls12/preeyanon/single_reads/line7i_single.fastq

##With quality trimming

Default minimum read length is 50.

#Assembly

All reads are assembled by Velvet and Oases with default parameters and k=21-31.

Raw reads were converted to fasta format prior to assembly (fastq-to-fasta.py in khmer/sandbox is used). 

#Reads mapping to genome

Reads were mapped to selected galGal4 sequences by Tophat2 with default parameters.
