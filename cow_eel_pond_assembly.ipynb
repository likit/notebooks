```python
op1 = open('Trinity.1000.fasta', 'w'); small_seq = 0
op2 = open('Trinity.3000.fasta', 'w'); medium_seq = 0
op3 = open('Trinity.large.fasta', 'w'); large_seq = 0
for seq in SeqIO.parse(gzip.open('Trinity.fasta.gz'), 'fasta'):
    if len(seq) < 1000:
        SeqIO.write(seq, op1, 'fasta')
        small_seq += 1
    elif (len(seq) > 1000 and len(seq) < 3000):
        SeqIO.write(seq, op2, 'fasta')
        medium_seq += 1
    else:
        SeqIO.write(seq, op3, 'fasta')
        large_seq += 1
print small_seq, medium_seq, large_seq
    
```

```python
families = {}
for seq in SeqIO.parse(gzip.open('Trinity.fasta.part.renamed.fasta.gz'), 'fasta'):
    fam = seq.id.split('.')[-1]
    families[fam] = families.get(fam, 0) + len(seq)
```

```python
fam_sizes = list(families.itervalues())
```

```python
len(fam_sizes) # number of families
```

```python
max(fam_sizes), min(fam_sizes)
```

```python
cumfreqs, lowlim, binsize, extrapoints = stats.cumfreq(fam_sizes,
                                                        max(fam_sizes))
norm_cumfreqs = cumfreqs / max(cumfreqs)
plot(norm_cumfreqs, '-', linewidth=1.5)
xlim(0,10000)
title('Cumulative Size of Transcript Families', size=18)
ylabel('Fraction', size=16)
xlabel('Size', size=16)

savefig('family_size_cdf.png')
```

```python
sorted_fam_sizes = sorted(families.keys(), key=lambda x: families[x], reverse=True)
```

```python
for s in sorted_fam_sizes[:10]:
    print s, families[s]
```

```python
for s in sorted_fam_sizes[:10]:
    seq_bags = []
    for seq in SeqIO.parse(gzip.open('Trinity.fasta.part.renamed.fasta.gz'), 'fasta'):
        fam = seq.id.split('.')[-1]
        if fam == s:
            seq_bags.append(seq)
    SeqIO.write(seq_bags, '%s.fa' % s, 'fasta')
```

```python
families_num = {}
for seq in SeqIO.parse(gzip.open('Trinity.fasta.part.renamed.fasta.gz'), 'fasta'):
    fam = seq.id.split('.')[-1]
    families_num[fam] = families_num.get(fam, 0) + 1
```

```python
_fam_nums = []
_fam_sizes = []
for f in families:
    _fam_nums.append(families_num[f])
    _fam_sizes.append(families[f])
    
plot(_fam_nums, _fam_sizes, '.')
yscale('log')
xscale('log')
title('Size vs Number of Transcript Family', size=18)
ylabel('Size(bp)', size=16)
xlabel('Number of Transcripts', size=16)
savefig('family_size_number.png')
```
